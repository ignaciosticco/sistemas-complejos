```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

1) a) Representación gráfica de las redes

Importación de datos

```python
G1=nx.read_edgelist('yeast_Y2H.txt')
G2=nx.read_edgelist('yeast_LIT.txt')
G3=nx.read_edgelist('yeast_AP-MS.txt')
```

Graficación

```python
nx.draw(G1, with_labels=False, font_weight='bold',node_size=10)
plt.show()
```

```python
nx.draw(G2, with_labels=False, font_weight='bold',node_size=10)
plt.show()
```

```python
nx.draw(G3, with_labels=False, font_weight='bold',node_size=10)
plt.show()
```

b) Tabla con características de las redes

Número total de nodos 

```python
nodos=[G1.number_of_nodes(),G2.number_of_nodes(),G3.number_of_nodes()]
print(nodos)
```

Número de enlaces

```python
enlaces=[G1.number_of_edges(),G2.number_of_edges(),G3.number_of_edges()]
print(enlaces)
```

c) Grado medio

```python
grado_medio = []
redes = [G1, G2, G3]

for red in range(1,4):
    medio = 0
    for i in range(0,nodos[red-1]):
        medio+=list(redes[red-1].degree)[i][1]
    grado_medio+=[float(medio)/nodos[red-1]]
    print(medio)

grado_medio        
#medio = medio / nodos
```
